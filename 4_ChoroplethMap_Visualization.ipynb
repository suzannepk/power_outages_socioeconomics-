## Choropleth Map of Temperatures across the US

This will teach you how to make a choropleth map from different data of the US with python's geopandas. 
A choropleth is a map that uses colors to shade different data based on its value.


```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import geopandas
```

```python
demo = pd.read_csv("data/CtyAvDemog2010.csv",delimiter=',', skiprows=0, low_memory=False, encoding='latin-1')
demo
```

```python
df = pd.read_csv("data/CtyAvTemp62016.csv", delimiter=',', skiprows=0, low_memory=False)
cityavgtmp_gdf = geopandas.GeoDataFrame(df)
cityavgtmp_gdf["TempFarenheit"] = 1.8*(cityavgtmp_gdf["TempK"]-273)+32
demo_gdf=geopandas.GeoDataFrame(pd.read_csv("CtyAvDemog2010.csv",delimiter=',', skiprows=0, low_memory=False, encoding='latin-1'))
merged_demo_cityavgtmp_gdf=demo_gdf.merge(cityavgtmp_gdf, how='inner', left_on="GEOID", right_on="GEOID")


merged_demo_cityavgtmp_gdf = geopandas.GeoDataFrame(
    merged_demo_cityavgtmp_gdf, 
    geometry=geopandas.points_from_xy(merged_demo_cityavgtmp_gdf["Lon_x"],
                                      merged_demo_cityavgtmp_gdf["Lat_x"]),
    crs="EPSG:4326"
)

merged_demo_cityavgtmp_gdf
```

```python
# making a choropleth map of us counties
uscounties_polygons_gdf = geopandas.read_file('https://raw.githubusercontent.com/holtzy/The-Python-Graph-Gallery/master/static/data/US-counties.geojson')
uscounties_polygons_gdf["id"] = uscounties_polygons_gdf["id"].astype(str).astype(np.int64)
uscounties_polygons_gdf = uscounties_polygons_gdf.rename(columns={"id":"FIPS_Code"})
uscounties_polygons_gdf
```

```python
# sjoin is one way to merge. This merges the two gdfs based on the geometry columns of the two gdfs
# and how the geometries intersect with each other
# (in the below case, the geometry in merged_medicare_cityavgtmp_gdf are latitude/longitude
# points, and in uscounties_polygons_gdf they are polygons representing us county shapes)
uscounties_gdf = uscounties_polygons_gdf.sjoin(merged_demo_cityavgtmp_gdf, how='inner', predicate="contains")
```
